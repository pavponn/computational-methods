### Зафиксируем полином пятой степени: 
$$f(x) = \frac{x^5}{10^5} - \frac{x^4}{10} + 10 x^3 - 20 x^2 + 12$$

```python
def f(x):
    return (1/10**5)*x**5 - (1/10)*x**4 + 10*x**3 - 20*x**2 + 12
```

```python
def f_p(x):
    return (5/10**5)*x**4 - (4/10)*x**3 + 30*x**2 - 40*x
```

```python
def f_pp(x):
    return (20/10**5)*x**3 - (12/10)*x**2 + 60*x - 40
```

# 1. Локализация корней

### Утверждение
Нет корней уравнения в промежутке $\left(-\infty; -1000 \right)$.
### Доказательство
Возьмем $x_L = -1000$. 

$f(x_L) < 0$. Пусть есть $x_0 < x_L : f(x_0) = 0$.

$    f'(x_L) > 0$, значит есть $x_1 : x_0 < x_1 < x_L \;,\;     f'(x_1) = 0$.

Дальнейшие строчки имеют тот же смысл как первая. В следующих производных высшего порядка нам важно только, что одна является производной другой.

$   f''(x_L) < 0$, значит есть $x_2 : x_1 < x_2 < x_L \;,\;    f''(x_2) = 0$.

$  f'''(x_L) > 0$, значит есть $x_3 : x_2 < x_3 < x_L \;,\;   f'''(x_3) = 0$.

$f^{IV}(x_L) < 0$, значит есть $x_4 : x_3 < x_4 < x_L \;,\; f^{IV}(x_4) = 0$.

$f^{IV}(x_4) = \frac{120}{10^5} x_4 - \frac{24}{10} = 0 \quad \Longrightarrow \quad x_4 = 2000 \notin \left(-\infty; -1000 \right)$.

Значит такой x_0 не существует.

# 2. Нахождение корней методом Ньютона

## Метод Ньютона

#### Достаточные условия
1. $|f' | \geq \alpha > 0$
2. $|f''| \leq \rho   < \infty$
3. $q \cdot |x_0 - x| < 1$, $q = {\frac{\rho}{2\alpha}}$

#### Условия остановки алгоритма
1. Относительная погрешность $err = \frac{\left|x^* - x\right|}{\left|x\right|} < 10^{-6} = eps$
1. Количество шагов $cnt\_steps > 10^{5} = step\_lim$

```python
def Newton_method(x_0):
    eps = 1e-6
    step_lim = 1e5
    
    x_prev = x_0
    x = x_0 - f(x_0)/f_p(x_0)
    cnt_steps = 0
    
    while not (abs(x - x_prev)/abs(x) < eps and cnt_steps > step_lim):
        x_prev = x
        x = x - f(x) / f_p(x)
        cnt_steps += 1
    
    return x
```

## Найдем $x_1 \in \left[-1; 0\right]$

```python
l,r = -1, 0
xs = np.linspace(l,r,1000)
ys = f(xs)
plt.axhline(0, color='gray')
plt.plot(xs,ys,'r')
plt.plot([  -1,   -1], [f(-1), f(0)], 'gray', linestyle = '--')
plt.plot([-0.4, -0.4], [f(-1), f(0)], 'gray', linestyle = '--')
plt.show()
```

$x_1 \in \left[-1; -0.4\right]$

$f'\left( -1\right) \approx 70.40005 \quad 
f'\left(-0.4\right) \approx 20.8256  \quad 
\Longrightarrow \quad |f'|  \geq 20.8 = \alpha$

$f''\left( -1\right) \approx -101.2002 \quad 
f''\left(-0.4\right) \approx -64.192   \quad 
\Longrightarrow \quad |f''| \leq 101.3 = \rho$

$q = {\frac{\rho}{2\alpha}} \approx 2.435$

Пусть $x_0 = -1$, тогда $q \cdot |x_0 - x| \leq q \cdot 0.4 \approx 0.974 < 1$

```python
print("f'(-1) = ", f_p(-1))
print("f'(-0.4) = ", f_p(-0.4))
alpha = 20.8
print("alpha = 20.8")
print("f''(-1) = ", f_pp(-1))
print("f''(-0.4) = ", f_pp(-0.4))
rho = 101.3
print("rho = 101.3")
q = rho/(2*alpha)
print("q*0.4 = ", q*0.4, " < 1")
```

```python
x_1 = Newton_method(-1)
x_1
```
